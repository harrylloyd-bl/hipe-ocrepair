```python
import copy
import glob
import json
```

### Split long records apart

```python
input_json = glob.glob("../data/raw/dta19_slim/de/*.jsonl")
input_json[0][:-6]
```

```python
for f in input_json:
    json_list = [json.loads(s) for s in open(f, encoding="utf8").readlines()]
    split_list = []
    for j in json_list:
        tu_split = j["ocr_hypothesis"]["transcription_unit"].split()
        if (len(tu_split) + 242) > 1600:
            print(j["document_metadata"]["document_id"])
            third = round(len(tu_split) / 3)
            p1, p2, p3 = copy.deepcopy(j), copy.deepcopy(j), copy.deepcopy(j)
            p1["ocr_hypothesis"]["transcription_unit"] = j["ocr_hypothesis"]["transcription_unit"][:third]
            p2["ocr_hypothesis"]["transcription_unit"] = j["ocr_hypothesis"]["transcription_unit"][third:third * 2]
            p3["ocr_hypothesis"]["transcription_unit"] = j["ocr_hypothesis"]["transcription_unit"][third * 2:]
            split_list.append(p1)
            split_list.append(p2)
            split_list.append(p3)
        elif (len(tu_split) + 242) > 1056:
            print(j["document_metadata"]["document_id"])
            half = round(len(tu_split) / 3)
            p1, p2, = copy.deepcopy(j), copy.deepcopy(j)
            p1["ocr_hypothesis"]["transcription_unit"] = j["ocr_hypothesis"]["transcription_unit"][:half]
            p2["ocr_hypothesis"]["transcription_unit"] = j["ocr_hypothesis"]["transcription_unit"][half:]
            split_list.append(p1)
            split_list.append(p2)
        else:
            split_list.append(j)

    with open(f[:-6] + "_slim.jsonl", "w") as f:
        for sj in split_list:
            f.write(json.dumps(sj) + "\n")
```
