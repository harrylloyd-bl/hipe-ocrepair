### JSON

```python
r0_json = {}
with open(os.path.join(trial_folder, "unmerged", 'qwendta0_r0.jsonl'), encoding="utf8") as f:
    for json_s in f.readlines():
        doc = json.loads(json_s)
        doc_id = doc["document_metadata"]["document_id"]
        r0_json[doc_id] = doc

r1_json = {}
with open(os.path.join(trial_folder, "unmerged", 'qwendta0_r1.jsonl'), encoding="utf8") as f:
    for json_s in f.readlines():
        doc = json.loads(json_s)
        doc_id = doc["document_metadata"]["document_id"]
        r1_json[doc_id] = doc

r2_json = {}
with open(os.path.join(trial_folder, "unmerged", 'qwendta0_r2.jsonl'), encoding="utf8") as f:
    for json_s in f.readlines():
        doc = json.loads(json_s)
        doc_id = doc["document_metadata"]["document_id"]
        r2_json[doc_id] = doc
```

```python
r1_df["document_id"].values.unique()
```

```python
for doc_id in r1_df["document_id"].values.unique():
    r0_json[doc_id] = r1_json[doc_id]

for doc_id in r2_df["document_id"].values.unique():
    r0_json[doc_id] = r2_json[doc_id]
```

```python
with open(os.path.join(trial_folder, 'qwendta0.jsonl'), "w", encoding="utf8", ) as f:
    for doc in r0_json.values():
        f.write(json.dumps(doc) + "\n")
```
